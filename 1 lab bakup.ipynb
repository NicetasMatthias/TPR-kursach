```python
from cvxopt.modeling import variable,op


start = 6         #начальные запасы на складе
s = [1,6,2,1,7,2] #спрос по дням
zp = 50           #цена завоза
hp = 8            #цена хранения


d1 = variable(1)
d2 = variable(1)
d3 = variable(1)
d4 = variable(1)
d5 = variable(1)
d6 = variable(1)

s1 = (start+d1>=s[0])                     #условия соответствия спросу
s2 = (start+d1-s[0]+d2>=s[1])
s3 = (start+d1-s[0]+d2-s[1]+d3>=s[2])
s4 = (start+d1-s[0]+d2-s[1]+d3-s[2]+d4>=s[3])
s5 = (start+d1-s[0]+d2-s[1]+d3-s[2]+d4-s[3]+d5>=s[4])
s6 = (start+d1-s[0]+d2-s[1]+d3-s[2]+d4-s[3]+d5-s[4]+d6>=s[5])
spros = [s1,s2,s3,s4,s5,s6]

dmz1 = (d1 >= 0) #условие неотрицательного привоза
dmz2 = (d2 >= 0)
dmz3 = (d3 >= 0)
dmz4 = (d4 >= 0)
dmz5 = (d5 >= 0)
dmz6 = (d6 >= 0)
dmz = [dmz1,dmz2,dmz3,dmz4,dmz5,dmz6]

dlf1 = (d1 <= 5)  #условие привоза не большего 5
dlf2 = (d2 <= 5)
dlf3 = (d3 <= 5)
dlf4 = (d4 <= 5)
dlf5 = (d5 <= 5)
dlf6 = (d6 <= 5)
dlf = [dlf1,dlf2,dlf3,dlf4,dlf5,dlf6]

ost3 = [(start+d1-s[0]+d2-s[1]+d3-s[2]+d4-s[3]+d5-s[4]+d6-s[5]==3)]

t1 = spros+dmz+dlf
t4 = t1+ost3


obj = (58*d1+50*d2+42*d3+34*d4+26*d5+18*d6+50*bool(d1)+50*bool(d2)+50*bool(d3)+50*bool(d3)+50*bool(d5)+50*bool(d6)+start*start*8-40*s[0]-32*s[1]-24*s[2]-16*s[3]-8*s[4])
#obj = (58*d1+50*d2+42*d3+34*d4+26*d5+18*d6+50*(bool(d1)+bool(d2)+bool(d3)+bool(d3)+bool(d5)+bool(d6))+start*start*8-40*s[0]-32*s[1]-24*s[2]-16*s[3]-8*s[4])
#obj = zp*((start+d1)+(start+d1-s[0]+d2)+(start+d1-s[0]+d2-s[1]+d3)+(start+d1-s[0]+d2-s[1]+d3-s[2]+d4)+(start+d1-s[0]+d2-s[1]+d3-s[2]+d4-s[3]+d5)+(start+d1-s[0]+d2-s[1]+d3-s[2]+d4-s[3]+d5-s[4]+d6))

print("1 задание")
lp1 = op(obj,t1)
lp1.solve(solver='glpk')
lp1.status

print(d1.value)
print(d2.value)
print(d3.value)
print(d4.value)
print(d5.value)
print(d6.value)
print(lp1.objective.value())



print("1 задание не glpk")
lp1 = op(obj,t1)
lp1.solve()
lp1.status

print(d1.value)
print(d2.value)
print(d3.value)
print(d4.value)
print(d5.value)
print(d6.value)
print(lp1.objective.value())



print("4 задание")
lp4 = op(obj,t4)
lp4.solve(solver='glpk')
lp4.status

print(d1.value)
print(d2.value)
print(d3.value)
print(d4.value)
print(d5.value)
print(d6.value)
print(lp4.objective.value())
print("-----------------------------------------------")
print(bool(d1.value))
print(bool(d2.value))
print(bool(d3.value))
print(bool(d4.value))
print(bool(d5.value))
print(bool(d6.value))
```

```python
from prettytable import PrettyTable
def f(det,dem,max_p,last,cdp,sp):
    demand = dem
    max_balance = 0 
    min_balance = 0 
    temp_max_balance = 0 
    temp_min_balance = 0 
    cache_balance = {} 
    cache_purchases = {} 
    counter = 0 
    result=[]
    flag = 0
    const_delivery_price = cdp
    var_delivery_price = 10
    storage_price = sp
    max_purchase = max_p
    last_balance = last
    cache_balance[last_balance] = []
    cache_balance[last_balance] = sp*last_balance
    max_balance = last_balance
    min_balance = last_balance

    for stage in reversed(range(len(demand))):
        for balance in range(stage*max_purchase+1):
            for purchases in range(max_purchase+1):
                if balance+purchases-demand[stage] > max_balance: 
                    break
                if balance+purchases >= demand[stage] and balance+purchases-demand[stage]>=min_balance:
                    if purchases != 0:
                        f_sum = (const_delivery_price+var_delivery_price*purchases) + storage_price * (balance+purchases-demand[stage]) + cache_balance[counter * 100 + balance + purchases - demand[stage]]
                    else:
                        f_sum = storage_price * (balance+purchases-demand[stage]) + cache_balance[counter * 100 + balance + purchases - demand[stage]]
                        
                    if (counter+1)*100+balance in cache_balance and f_sum < cache_balance[(counter + 1) * 100 + balance]: 
                        cache_balance[(counter + 1) * 100 + balance] = f_sum
                        cache_purchases[(counter + 1) * 100 + balance] = purchases
                        if flag == 0:
                            temp_min_balance=balance
                            flag = 1
                        if balance > temp_max_balance:
                            temp_max_balance=balance
                    elif (counter+1)*100+balance not in cache_balance:
                        cache_balance[(counter + 1) * 100 + balance] = []
                        cache_balance[(counter + 1) * 100 + balance]=f_sum
                        cache_purchases[(counter + 1) * 100 + balance] = []
                        cache_purchases[(counter + 1) * 100 + balance] = purchases
                        if flag == 0:
                            temp_min_balance=balance
                            flag = 1
                        if balance > temp_max_balance:
                            temp_max_balance = balance
        counter += 1
        max_balance = temp_max_balance
        min_balance = temp_min_balance
        temp_min_balance = 0
        temp_max_balance = 0
        flag=0

    index_search = 0
    new_index_search = 0
    storage_cost = 0
    delivery_cost = 0
    if det!= 0:
        print("Итоговые затраты: ",cache_balance[counter*100])

    for stage in reversed(range(1,len(demand)+1)): # Вывод результатов
        index_search = cache_purchases[100*stage+new_index_search]
        if index_search !=0:
            delivery_cost = const_delivery_price + var_delivery_price*index_search
        else:
            delivery_cost=0
        new_index_search += index_search-demand[len(demand)-stage]
        storage_cost = new_index_search * storage_price
        result.append(index_search)
        if det!=0:
            print()
            print(len(demand)-stage+1,'месяц')
            print('Закупка: ',index_search)
            print('Стоимость хранения: ',storage_cost)
            print('Затраты на доставку: ',delivery_cost)
            print('Итого за месяц: ',delivery_cost + storage_cost,)
    if det!=0:
        print()
        print('Итоговый план:')
        print(result)
    return result


print('Задание №1')
det = [1,6,2,1,7,2] # спрос
mp = 5 #максимальная закупка в месяц
b = 0 #остаток на конец периода
cdp = 50 #постоянные затраты на доставку
sp = 8 #цена хранения

d = 400 #интервал в пределах которого ищется верхняя граница

res1 = f(1,det,mp,b,cdp,sp)
print('-'*50)
print('Задание №2')
    
max_sp = sp+1
ressmax = f(0,det,mp,b,cdp,max_sp)

while res1 == ressmax :
    max_sp = max_sp + 1
    ressmax = f(0,det,mp,b,cdp,max_sp)
    #print('верхняя граница: ',min_sp,' план: ',ressmax)
    if max_sp >= sp+d:
        break
if ressmax == res1:
    print('При поиске вехрней границы затрат на хранение вышли за допустимый интервал')
else:
    max_sp = max_sp - 1
    print('Верхняя граница затрат на хранение: ',max_sp)


min_sp = sp-1
ressmin = f(0,det,mp,b,cdp,min_sp)

while res1 == ressmin :
    min_sp = min_sp - 1
    ressmin = f(0,det,mp,b,cdp,min_sp)
    #print('нижняя граница: ',min_sp,' план: ',ressmin)
    if min_sp <= 0:
        break
if ressmin == res1:
    print('При поиске нижней границы затрат на хранение вышли за допустимый интервал (<=0)')
else:
    min_sp = min_sp + 1
    print('Нижняя граница затрат на хранение: ',min_sp)


print('-'*50)
print('Задание №3')

max_cdp = cdp+1
resdmax = f(0,det,mp,b,max_cdp,sp)

while res1 == resdmax :
    max_cdp = max_cdp + 1
    resdmax = f(0,det,mp,b,max_cdp,sp)
    #print('верхняя граница: ',max_cdp,' план: ',resdmax)
    if max_cdp >= cdp+d:
        break
if resdmax == res1:
    print('При поиске верхней границы постоянных затрат на доставку вышли за допустимый интервал')
else:   
    max_cdp = max_cdp - 1
    print('Верняя граница постоянных затрат на доставку: ',max_cdp)



min_cdp = cdp-1
resdmin = f(0,det,mp,b,min_cdp,sp)

while res1 == resdmin :
    min_cdp = min_cdp - 1
    resdmin = f(0,det,mp,b,min_cdp,sp)
    #print('нижняя граница: ',min_cdp,' план: ',resdmin)
    if min_cdp <= 0:
        break
if resdmin == res1:
    print('При поиске нижней границы постоянных затрат на доставку вышли за допустимый интервал (<=0)')
else:
    min_cdp = min_cdp + 1
    print('Нижняя граница постоянных затрат на доставку: ',min_cdp)


print('-'*10)
print('Задание №4')
res4 = f (1,det,mp,3,cdp,sp)
```
